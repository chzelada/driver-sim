```python
import googlemaps
from dotenv import load_dotenv
from googlemaps import convert
import os
```

```python
load_dotenv()
api_key = os.getenv('google_map_key')
gmaps = googlemaps.Client(key=api_key)
```

```python
route1 = gmaps.directions(origin=[14.57341656112588, -90.52457302750229],destination=[14.60804486286022, -90.48687551400758])
```

```python
route_poly = route1[0]['overview_polyline']['points']
```

```python
route_points = convert.decode_polyline(route_poly)
display(route_points)
```

```python
# converto route_point into a pandas dataframe
import pandas as pd
df = pd.DataFrame(route_points, columns=['lat', 'lng'])
df.head()
```
